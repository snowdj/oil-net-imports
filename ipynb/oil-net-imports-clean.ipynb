```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")

% matplotlib inline
```

```python
imports_weekly_df = pd.read_excel('../data/PET_MOVE_WKLY_DC_NUS-Z00_MBBLPD_W.xls',
                                  sheetname='Data 3', header=2)
imports_weekly_df.columns = ['date','net','crude','petroleum']
pd.to_datetime(imports_weekly_df['date'])
imports_weekly_df.head()
```

```python
imports_weekly_df.set_index('date').iloc[561:].plot(figsize=(15,10))
plt.axvline('2017-01-01', c='purple', label='OPEC Production Cut Start')
plt.legend()
plt.title('Weekly Crude and Petroleum Products Imports')
plt.xlabel('Year')
plt.ylabel('BB/day in Thousands')
plt.savefig('../assets/prod_forever.png')
plt.show()
```
